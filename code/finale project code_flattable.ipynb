```python
##############################################################################
##########                DATABASE FUNCTIONS                     #############
##############################################################################

#### List the name of the tables on the database 
def listTables():
    import sqlite3 as sql3
    db = sql3.connect("proj.db")
    cur = db.cursor()
    cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
    res = cur.fetchall()
    cur.close()
    db.close()
    return(res)

#### Read function to import data from the SQL to a pandas dataframe.
def readSQL(query):
    import pandas as pd
    import sqlite3 as sql3
    db = sql3.connect("proj.db")
    df = pd.read_sql_query(query, db)
    db.close()
    return(df)

#### Write a pandas dataframe into an SQL table. Use overwrite=True if you want to delete 
#### first a pre-existent table with the same name. Use append=True if you want to append
#### the data in the dataframe to a pre-existent table.
```

```python
def make_one_hot_time(data,col_name, ncat):
        data[col_name+'-'+'breakfast'] = 0 
        data[col_name+'-'+'lunch'] = 0 
        data[col_name+'-'+'dinner'] = 0 
        data[col_name+'-'+'latenight'] = 0 
        data.loc[('06:00:00' <= data[col_name].apply(str)) & (data[col_name].apply(str) <'11:00:00') ,col_name+'-'+'breakfast'] = 1
        data.loc[('11:00:00' <= data[col_name].apply(str)) & (data[col_name].apply(str) <'17:00:00') ,col_name+'-'+'lunch'] = 1
        data.loc[('17:00:00' <= data[col_name].apply(str)) & (data[col_name].apply(str) <'22:00:00') ,col_name+'-'+'dinner'] = 1    
        data.loc[('22:00:00' <= data[col_name].apply(str)) | (data[col_name].apply(str) <'06:00:00') ,col_name+'-'+'latenight'] = 1
```

```python
def make_one_hot_weekday(data,col_name, ncat):
    for i in range(1,ncat+1):
        data[col_name+'-'+str(i)] = None
        data.loc[(data[col_name] == i),col_name+'-'+str(i)] = 1
        data.loc[(data[col_name] != i),col_name+'-'+str(i)] = 0 
```

```python
def writeSQL(df,tablename,overwrite=False, append=False):
    import pandas as pd
    import sqlite3 as sql
    db = sql.connect(DB_FILE)
    if (overwrite):
        action = "replace"
    elif (append):
        action = "append"
    else: 
        action = "fail"
    df.to_sql(tablename, db, if_exists=action)
    db.close()
    
writeSQL(flat, "flat")
```
